<script>
function predict() {
    const url = "http://localhost:8587/api/titanic/predict";
    const responseDiv = document.getElementById('response');

    const data = {
        pclass: parseInt(document.getElementById('pclass').value),
        sex: document.getElementById('sex').value,
        age: parseFloat(document.getElementById('age').value),
        sibsp: parseInt(document.getElementById('sibsp').value),
        parch: parseInt(document.getElementById('parch').value),
        fare: parseFloat(document.getElementById('fare').value),
        embarked: document.getElementById('embarked').value,
        alone: document.getElementById('alone').checked,
        name: "User"
    };

    fetch(url, {
        method: 'POST',
        headers: { 'Content-Type': 'application/json' },
        body: JSON.stringify(data)
    })
    .then(response => response.json())
    .then(result => {
        const survive = (result.survive * 100).toFixed(2);
        const die = (result.die * 100).toFixed(2);
        responseDiv.innerHTML = `Survival: ${survive}% <br> Death: ${die}%`;
        responseDiv.style.color = result.survive > 0.5 ? "green" : "red";
    })
    .catch(error => {
        console.error('Error:', error);
        responseDiv.innerHTML = "Error contacting API. Ensure Flask is running.";
        responseDiv.style.color = "red";
    });
}
</script>
